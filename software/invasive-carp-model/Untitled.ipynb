```python
import pandas as pd
```

```python
data = pd.read_csv("regions_data.csv")
```

```python
data
```

```python
for index, row in data.iterrows():
    year = row["Year"]  # Extract the year
    region_data = {col: row[col] for col in data.columns if col != "Year"}  # Extract region data (Y/N)
    
    # Print the extracted data
    print(f"Year: {year}")
    for region, value in region_data.items():
        print(f"  {region}: {value}")
```
